```python
from sklearn.decomposition import PCA
import pandas as pd
import  matplotlib.pyplot as pl
import numpy as np
```

READ DATA

```python

data = pd.read_csv("pca_dataset.txt",sep = " ",header=None)

pca = PCA(2)
pca.fit(data)

print("Contained energy of two components =",pca.explained_variance_)

pl.scatter(x=data[0], y= data[1], marker = "*", alpha=0.5,  label="samples")
for i in range(2):
    to_plot = pca.components_[i] * pca.explained_variance_[i]
    pl.plot([0,to_plot[0]], [0,to_plot[1]],color= "C4")
```

Load Image of Racoon

```python
import scipy
import cv2
rakoon = scipy.misc.face(gray=True)
rakoon = cv2.resize(rakoon,(249,185))
pl.imshow(rakoon)
```

Image with All, 120, 50, 10 Principal Components

```python
list_of_numbers_of_components = [185,120,50,10]

for noc in list_of_numbers_of_components:
    print("Image with",noc, "components")
    pca = PCA(noc)
    pca.fit(rakoon)
    new_rakoon = pca.inverse_transform(pca.transform(rakoon))
    pl.imshow(new_rakoon)
    pl.show()
```

At what number is the “energy” lost through truncation smaller than 1%?

```python
pca = PCA(0.99)
pca.fit(rakoon)
print("Number of components needed for less than 1% loss is: ",len(pca.components_))
```

Pedestrian Data Read

```python
data = pd.read_csv("data_DMAP_PCA_vadere.txt",sep = " ",header=None)
data.head()
```

Visualizing first two pedestrians' path

```python
x_first, y_first, x_second, y_second =data[0], data[1], data[2], data[3]
pl.plot(x_first,y_first, color = "C3")
pl.plot(x_second,y_second, color = "C4")
pl.xticks(np.arange(0, 25, 2.0))
pl.yticks(np.arange(0, 25, 2.0))
pl.show()
```

PCA of 30-dimensional data

```python
pca = PCA(30)
pca.fit(data) 
print("Explained Variance of all components:", pca.explained_variance_)
print("Two Components Contained Percentage: ", sum(pca.explained_variance_[:2])/sum(pca.explained_variance_))
print("Two Components Contained Percentage: ", sum(pca.explained_variance_[:3])/sum(pca.explained_variance_))
```
