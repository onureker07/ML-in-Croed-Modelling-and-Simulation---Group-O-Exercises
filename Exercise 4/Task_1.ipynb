## Import the necessary libraries

```python
import numpy as np
import matplotlib.pyplot as plt
```

## Generate phase portraits

```python
def plot_phase_portrait(A):
    """ Plot the phase portraits of the given matrix A.

    Args:
        A (list): Parameterized 2 by 2 matrix.
    """
    
    x = np.linspace(-2, 2, 20)
    y = np.linspace(-2, 2, 20)
    
    X, Y = np.meshgrid(x, y)
    
    dphidt = A[0][0]*X + A[0][1]*Y
    dphidx = A[1][0]*X + A[1][1]*Y
    
    dphidt = dphidt / np.sqrt(dphidt**2 + dphidx**2)
    dphidx = dphidx / np.sqrt(dphidt**2 + dphidx**2)
    
    #Calculate the eigenvalues of the matrix
    eigenvalues = np.linalg.eigvals(A)

    plt.title('alpha = '+str(p)+" lambda1 = "+str(eigenvalues[0])+" lambda2 = "+str(eigenvalues[1]))
    plt.quiver(X, Y, dphidt, dphidx)
    plt.show()
```

```python
# Stable node
p = -1
A = [[-2,0],
     [0,p]]
plot_phase_portrait(A)
```

```python
# Stable focus
p = -0.5
A = [[p,-1],
     [1,p]]
plot_phase_portrait(A)
```

```python
# Unstable focus
p = 0.5
A = [[p,-1],
     [1,p]]
plot_phase_portrait(A)
```

```python
# Unstable node
p = 1
A = [[2,0],
     [0,p]]
plot_phase_portrait(A)
```

```python
# Unstable saddle
p = -1
A = [[2,0],
     [0,p]]
plot_phase_portrait(A)
```
