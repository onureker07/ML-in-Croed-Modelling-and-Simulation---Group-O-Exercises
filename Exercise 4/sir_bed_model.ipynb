```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

from scipy.integrate import solve_ivp

from sir_model import *
```

```python
A, d, nu, mu0, mu1, beta =20, 0.1, 1, 10, 10.45, 11.5
bs = [0.01, 0.022, 0.03]
sir = [[195.3,0.052,4.4], [195.7,0.03,3.92], [193,0.08,6.21]]
for b in bs:
    for sir_0 in sir:
        result = solve_ivp(model, [0, 100], sir_0, args=( mu0, mu1, beta,A,d,nu,b), t_eval=np.linspace(0, 100, 1000))

        t = result.t
        S = result.y[0]
        I = result.y[1]
        R = result.y[2]

        
        
        #Scatter S vs I with color map R and indicate that R is the color in the plot
        

        plt.scatter(S, I, c=R, label = 'R')
        plt.colorbar().set_label('R values')
        plt.xlabel('S')
        plt.ylabel('I')
        plt.title('SIR_0 = '+str(sir_0)+" and b = "+str(b))
        plt.show()
```
